```python
import pyodbc
import pandas as pd

# Ganti informasi koneksi sesuai dengan database Anda
server = 'ID-LTP-AFI-0028'
database = 'Northwind'
# Hapus bagian username dan password
conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes'

# Membuat objek koneksi
conn = pyodbc.connect(conn_str)

# Membuat objek cursor untuk eksekusi perintah SQL
cursor = conn.cursor()

# Contoh eksekusi query SELECT
query_category = 'select E.CategoryName,Sum(E.UnitPrice*E.Quantity-E.Discount) as Amount_Transaction ,Sum(E.Discount) As Total_Discount from (Select C.*,D.CategoryName From (SELECT B.CategoryID,A.* FROM Order_Details A Join Products B on A.ProductID=B.ProductID) C join Categories D on C.CategoryID = D.CategoryID) E group by E.CategoryName Order by Amount_Transaction DESC'
cursor.execute(query_category)

# Mendapatkan hasil query
rows = cursor.fetchall()

# Mengubah hasil query menjadi DataFrame
df_category = pd.DataFrame([tuple(row) for row in rows], columns=[column[0] for column in cursor.description])

dataset_category = df_category

#dataset_category.head(20)


from pivottablejs import pivot_ui
pivot_ui(dataset_category)

```
